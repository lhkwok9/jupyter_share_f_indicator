```python
#Use the relative Strength Index (RSI) and Python to determine 
#if a stock is being-over bought or over-sold
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('stock_dfs_handmade/aapl.csv')
df
```

```python
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
df
```

```python
%matplotlib notebook
plt.figure(figsize=(12.2,4.5))
plt.plot(df.index, df['Adj Close'])
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Adj CLose Price', fontsize = 18)
plt.show()
```

```python
#Prepare data to caculate the RSI

#Get the diff in p
delta = df['Adj Close'].diff(1)
delta
```

```python
#get rid of NAN
delta = delta.dropna()
delta
```

```python
#Get up and down
up = delta.copy()
down = delta.copy()

up[up<0] = 0
down[down>0] = 0
```

```python
#Get the time period
period = 14
#Calculate the average gain and average loss
AVG_Gain = up.rolling(window=period).mean()
AVG_Loss = abs(down.rolling(window=period).mean())
```

```python
#Cal the RSI

#Cal the RS
RS = AVG_Gain / AVG_Loss
#Cal the RSI
RSI = 100.0 - (100.0 / (1.0 + RS))
RS
```

```python
#Show RSI
plt.figure(figsize=(12.2,4.5))
RSI.plot()
plt.show()
```

```python
#put all tgt

new_df = pd.DataFrame()
new_df['Adj Close'] = df['Adj Close']
new_df['RSI'] = RSI
new_df
```

```python
%matplotlib notebook
plt.figure(figsize=(12.2,4.5))
plt.plot(new_df.index, new_df['Adj Close'])
plt.legend(new_df.columns.values, loc = 'upper left')
plt.show()

plt.figure(figsize=(12.2,4.5))
plt.plot(new_df.index, new_df['RSI'])
plt.axhline(0, linestyle='--', alpha = 0.5, color = 'red')
plt.axhline(10, linestyle='--', alpha = 0.5, color = 'orange')
plt.axhline(20, linestyle='--', alpha = 0.5, color = 'yellow')
plt.axhline(30, linestyle='--', alpha = 0.5, color = 'grey')
plt.axhline(70, linestyle='--', alpha = 0.5, color = 'grey')
plt.axhline(80, linestyle='--', alpha = 0.5, color = 'yellow')
plt.axhline(90, linestyle='--', alpha = 0.5, color = 'orange')
plt.axhline(100, linestyle='--', alpha = 0.5, color = 'red')
```
